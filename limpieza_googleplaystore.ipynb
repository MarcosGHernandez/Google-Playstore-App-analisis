```python
#Carga de librerias(dependencias previamente instaladas)
import pandas as pd
import numpy as np
```

```python
# Cargar el dataset
df = pd.read_csv('googleplaystore.csv')

# Eliminación de fila corrupta detectada 
df = df.drop(10472).reset_index(drop=True)
```

```python
# Normalización de datos 

# Eliminar '+' y ',' de Installs y convertir a entero
df['Installs'] = df['Installs'].astype(str).str.replace('+', '').str.replace(',', '')
df['Installs'] = pd.to_numeric(df['Installs'])

# Eliminar '$' de Price y convertir a float
df['Price'] = df['Price'].astype(str).str.replace('$', '')
df['Price'] = pd.to_numeric(df['Price'])

# Normalizacion de Size
def clean_size(size):
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024 # Convertir KB a MB
    elif 'Varies with device' in size:
        return np.nan
    else:
        return np.nan

df['Size'] = df['Size'].map(clean_size)
```

```python
# Tratamiento de Nulos en Rating 
# Rellenar nulos con la mediana
rating_median = df['Rating'].median()
df['Rating'].fillna(rating_median, inplace=True)

# Eliminación de duplicados
filas_antes = len(df)
df.drop_duplicates(inplace=True)
filas_despues = len(df)
print(f"Se eliminaron {filas_antes - filas_despues} filas duplicadas.")
```

```python
# Guardado
nombre_archivo_final = 'googleplaystore_cleaned.csv'
df.to_csv(nombre_archivo_final, index=False)

print(f"\n¡Limpieza completada! El archivo se ha guardado como: {nombre_archivo_final}")
print("\nInformación del Dataset Limpio:")
print(df.info())
```
